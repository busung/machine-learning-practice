```python
import os
file_path = '/content/drive/MyDrive/산데과'
file_list = os.listdir(file_path)
iris = pd.read_csv(file_path + '/' + file_list[0])
iris.head()
```

```python
flower_name = iris['Species'].value_counts().index
flower_name
```

```python
seto = iris[iris['Species']==flower_name[0]]
vir = iris[iris['Species']==flower_name[1]]
ver = iris[iris['Species']==flower_name[2]]
seto.head()
```

```python
list_ = [seto,vir,ver]
col_list = list(iris.columns[[1,2,4]])
color = ['red','skyblue','gray']
```

```python
for i in col_list:
  for j in range(len(list_)):
      sns.distplot(list_[j][i],bins=10,color=color[j])
      #red = seto,skyblue = vir,gray = ver
  labels = ['setosa','virginica','versicolor']
  plt.legend(labels)
  plt.show()
```

```python
mean_list=[]
for i in col_list:
  temp = []
  for j in list_:
    temp.append(j[i].mean())
  mean_list.append(temp)
mean_list
```

```python
for i in range(len(col_list)):
  for j in range(len(list_)):
      sns.distplot(list_[j][col_list[i]],bins=10,color=color[j])
      #red = seto,skyblue = vir,gray = ver
  for m in mean_list[i]:
    plt.axvline(x=m,color='black',linewidth=1)
  labels = ['setosa','virginica','versicolor']
  plt.legend(labels)
  plt.show()
```
