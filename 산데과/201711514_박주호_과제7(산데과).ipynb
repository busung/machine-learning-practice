```python
total_dt = pd.concat([test_dt,train_dt])
total_dt.head()
x=total_dt.iloc[:,1:18]
x.head()
```

```python
import sklearn.preprocessing
scaler = sklearn.preprocessing.MinMaxScaler()
x=scaler.fit_transform(x)
```

```python
x=pd.DataFrame(x)
x
```

```python
label=total_dt.iloc[:,-1]
x["label"]=list(label)
test_dt = x.iloc[0:9184,]
train_dt = x.iloc[9184:,]
```

```python
train_dt.sort_values(['label']).reset_index(drop=True)
label_train=[]
for i in range(len(train_dt)):
  if i < int(np.round(len(train_dt)*0.33)):
    label_train.append(0)
  elif i >= int(np.round(len(train_dt)*0.33)) and i< int(np.round(len(train_dt)*0.66)):
    label_train.append(1)
  else:
    label_train.append(2)

pd.Series(label_train).value_counts()
```

```python
train_dt = train_dt.iloc[:,0:17]
train_dt
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(train_dt,label_train,test_size=0.3)
print(len(train_x))
print(len(test_x))
print(len(train_y))
print(len(test_y))
```

# model 생성

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

```python
model = RandomForestClassifier()
model = model.fit(train_x,train_y)
pred_test = model.predict(test_x) 
m = confusion_matrix(test_y,pred_test)
acc = np.sum(np.diag(m)/len(test_y))
print("{:0.2f}".format(acc*100))
```

```python
model = RandomForestClassifier()
model = model.fit(train_dt,label_train)
pred_test = model.predict(test_x) 
m = confusion_matrix(test_y,pred_test)
acc = np.sum(np.diag(m)/len(test_y))
print("{:0.2f}".format(acc*100))
```

```python
model.feature_importances_
fi = pd.concat([pd.DataFrame(list(train_x.columns)),pd.DataFrame(model.feature_importances_)],axis=1)
fi.columns = ["columns","importances"]
print(fi)
fi.sort_values("importances")
```

```python
res = []
for i in range(1,11):
  model = RandomForestClassifier(n_estimators = i*100)#tree의 개수
  model = model.fit(train_x,train_y)
  pred_test = model.predict(test_x)
  m = confusion_matrix(test_y,pred_test)
  acc = np.sum(np.diag(m)/len(test_y))
  res.append(acc)
  print("{:0.2f} number{}".format(acc*100,i*100))
  model.feature_importances_
  fi = pd.concat([pd.DataFrame(list(train_x.columns)),pd.DataFrame(model.feature_importances_)],axis=1)
  fi.columns = ["columns","importances"]
  fi = fi.sort_values("importances")
  print(fi)
```

```python
#900개의 트리에서 상위 10개의 변수 2 11 1 10 4 8 9 3 7 6 15만을 선택하여 모델 생성
train_x_2 = train_x.loc[:,[16, 5, 11, 3, 10, 8, 9, 4, 7, 6, 15]]
test_x_2 = test_x.loc[:,[16, 5, 11, 3, 10, 8, 9, 4, 7, 6, 15]]
model = RandomForestClassifier(n_estimators = 8*100)#tree의 개수
model = model.fit(train_x_2,train_y)
pred_test = model.predict(test_x_2)
m = confusion_matrix(test_y,pred_test)
acc = np.sum(np.diag(m)/len(test_y))
print("{:0.2f}".format(acc*100))
```
